```python
# ---------------------------------------------------------------
# DATA GENERATION FOR SUPERMARKET TRANSACTIONS
# Contributor: Abdiqalaq
# ---------------------------------------------------------------

import random
import pandas as pd

# ---------------------------------------------------------------
# Step 1: Create a pool of at least 30 supermarket items
# ---------------------------------------------------------------
item_pool = [
    "Bread", "Milk", "Eggs", "Butter", "Cheese", "Apples", "Bananas", "Chicken",
    "Rice", "Pasta", "Tomatoes", "Onions", "Cereal", "Yogurt", "Juice", "Beef",
    "Fish", "Salt", "Sugar", "Tea", "Coffee", "Chips", "Soda", "Water",
    "Flour", "Oil", "Spices", "Ice Cream", "Chocolate", "Lettuce"
]

# ---------------------------------------------------------------
# Step 2: Generate at least 3000 transactions
# Each transaction will have 2â€“7 randomly selected items
# ---------------------------------------------------------------
num_transactions = 3000   # Requirement: minimum 3000
transactions = []

for _ in range(num_transactions):
    transaction_size = random.randint(2, 7)           # number of items in one transaction
    transaction = random.sample(item_pool, transaction_size)  
    transactions.append(transaction)

# ---------------------------------------------------------------
# Step 3: Save as CSV for later use in Apriori
# ---------------------------------------------------------------
df = pd.DataFrame({"Transaction": transactions})
df.to_csv("transactions.csv", index=False)

print("Generated 3000 supermarket transactions successfully!")
```
